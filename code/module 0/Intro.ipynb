# LangChain

<p>The aim of LangChain is to make it <u>easy to build</u> LLM applications. One type of LLM application we can build is an <span style="color:red">agent</span>. There’s a lot of excitement around building agents because they can <b>automate a wide range of tasks</b> that were previously impossible.</p>

<p>In practice though, it is incredibly <b>difficult</b> to build systems that <u>reliably execute</u> on these tasks. As LangChain worked with their users to put agents into production, they’ve learned that <u>more control</u> is often necessary. For e.g., we might need an agent to always call a specific tool first or use different prompts based on its state.</p>

<p>To tackle this problem, LangChain has built <a href="https://docs.langchain.com/oss/python/langgraph/overview" target="_blank">LangGraph</a> — <span style="color:red">a framework for building agent and multi-agent applications</span>. Separate from the LangChain package, LangGraph’s core design philosophy is to <b>help developers add better <u>precision and control</u> into agent workflows</b>, suitable for the complexity of real-world systems.</p>

## Chat models

<p>Chat Models take a sequence of messages as input and return messages as output. LangChain supports many models via <a href="https://docs.langchain.com/oss/python/integrations/chat" target="_blank">third-party integrations</a>.</p>

<p>In this course, we will use the <a href="https://docs.langchain.com/oss/python/integrations/chat/google_generative_ai" target="_blank">ChatGoogleGenerativeAI</a> integration.</p>

<p>There are a few standard parameters that we can set with chat models. Two of the most common are: <ul><li><b>model:</b> the name of the model</li> <li><b>temperature:</b> the sampling temperature</li></ul></p>

<p><i><b>Note:</b> Temperature controls the randomness or creativity of the model's output where low temperature (close to 0) is more deterministic and focused outputs. This is good for tasks requiring accuracy or factual responses. High temperature (close to 1) is good for creative tasks or generating varied responses.</i></p>

```python
import dotenv

dotenv.load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

gemini_25_fl = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", temperature=0)
gemini_35_fl = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite-preview", temperature=0)
```

<p>Chat models in LangChain have a number of <a href="https://reference.langchain.com/python/langchain-core/runnables" target="_blank">default methods</a>. For the most part, we'll be using:<ul><li><b><a href="https://docs.langchain.com/oss/python/langchain/models#stream" target="_blank">stream:</a></b> stream back chunks of the response</li><li><b><a href="https://docs.langchain.com/oss/python/langchain/models#invoke" target="_blank">invoke:</a></b> call the chain on an input</li></ul></p>

<p><i><b>Note:</b> Chat models take <a href="https://docs.langchain.com/oss/python/langchain/messages" target="_blank">messages</a> as input. Messages have a role (that describes who is saying the message) and a content property.</i></p>

```python
from langchain_core.messages import HumanMessage

# Create a message
msg = HumanMessage(content="Hello world", name="Dibyajyoti")

# Message list
messages = [msg]

# Invoke the model with a list of messages 
gemini_25_fl.invoke(messages)
```

<p>We get an AIMessage response. Also, note that we can just invoke a chat model with a string. When a string is passed in as input, it is converted to a <b>HumanMessage</b> and then passed to the underlying model.</p>

```python
gemini_25_fl.invoke("hello world")
```

```python
gemini_35_fl.invoke("hello world")
```

<p>The interface is consistent across all chat models and models are typically initialized once at the start up each notebooks.</p>

<p>So, we can easily switch between models without changing the downstream code if you have strong preference for another provider.</p>
